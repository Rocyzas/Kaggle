```python
# We dont need id, language and lang_abv, since it does not provide supporting information to the model. Therefore remove them.
train_df = train[['premise', 'hypothesis']]
train_label = train[['label']]
test_df = test[['premise', 'hypothesis']]
```

```python
# We can use translator API to translate all sentences to English, or use multilanguage tokenisors of HuggingFace.
df = pd.DataFrame(train_df)
df['t1'] = df.apply(lambda row: nltk.word_tokenize(row['premise']), axis=1)
df['t2'] = df.apply(lambda row: nltk.word_tokenize(row['hypothesis']), axis=1)
train_df = df
```
