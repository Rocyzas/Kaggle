```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# to make matplotlib plots visible in jupyter notebook
%matplotlib inline

# to auto reload page
%reload_ext autoreload
%autoreload 2
```

# Simple PLotting

```python
x_1 = np.linspace(0, 5, 10)
y_1 = x_1**2
plt.plot(x_1, y_1)
plt.title('Tittle')

# adding labels
plt.xlabel('x lab')
plt.ylabel('y lab')

plt.show()
```

# Multiple Plots

```python
plt.subplot(1, 2, 1)
plt.plot(x_1, y_1, 'r')

plt.subplot(1, 2, 2)
plt.plot(x_1, y_1, 'b')
```

# Figure

```python
fig_1 = plt.figure(figsize=(6,4), dpi=100)

# left bottom width height
axes1 = fig_1.add_axes([0, 0, 1, 1])
axes1.set_xlabel('Int')
axes1.set_ylabel('Squared')
axes1.set_title('Squared chart')
axes1.plot(x_1, y_1, label='x/x^2')
axes1.plot(y_1, x_1, 'r', label='x^2/x')
axes1.legend(loc=0)

axes2 = fig_1.add_axes([0.45, 0.45, 0.4, 0.3])
axes2.set_xlabel('Int')
axes2.set_ylabel('Squared')
axes2.set_title('Squared chart')
axes2.plot(x_1, y_1, 'g')
```

# Subplots

```python
fig2, axes2 = plt.subplots(figsize=(8,4), nrows=1, ncols=3)
# space between polots
plt.tight_layout()
# accessing axe by index
axes2[2].set_title('plot3')
axes2[2].set_xlabel('X')
axes2[2].set_ylabel('X squared')
axes2[2].plot(x_1, y_1)
```

# Advanced options

```python
fig3 = plt.figure(figsize=(6, 4))
axes3 = fig3.add_axes([0,0,1,1])
axes3.plot(x_1, y_1, color = 'navy', 
            alpha=0.7, lw=2, ls='-.', label="asdasd",
            marker='o', markersize=7, markerfacecolor='y',
            markeredgecolor='y', markeredgewidth=4)
axes3.legend(loc=0)

# Set the lower and upper bound of x & y axis
axes3.set_xlim([0,3])
axes3.set_ylim([0,8])


plt.show()
```
